**Introducción al Data Set: *'IMDB movies dataset'***

Este **conjunto de datos IMDB** explora más de 10.000 películas de todo el mundo.\
Contiene información sobre películas, incluidos sus nombres, fechas de estreno, valoraciones de usuarios, géneros, descripciones generales, miembros del reparto y del equipo, títulos originales, estado de producción, idiomas originales, presupuestos, ingresos y países de origen.\
Estos datos se pueden utilizar para diversos análisis, como identificar tendencias en géneros cinematográficos, explorar la relación entre presupuesto e ingresos y predecir el éxito de películas futuras.
<br>\
**Diccionario de Datos (*Variables*)**

- **names (*Nombres*):** Nombre de la película.\
Variable que contiene el título de la película.\
Este campo es útil para identificar de manera única cada película en el conjunto de datos.

- **date_x (*Fecha*):** Fecha de lanzamiento.\
Fecha en la que la película fue estrenada.\
Variable permite realizar análisis relacionados con la cronología de lanzamientos de películas.

- **score (*Puntuación*):** Puntuación de la película.\
Representa la calificación o puntaje asignado a la película.\
Este puntaje puede ser una medida de la calidad de la película, como una calificación de audiencia o una calificación de críticos.

- **genre (*Género*):** Géneros de la película.\
Variable que enumera los géneros cinematográficos a los que pertenece la película.\
Puede haber varios géneros separados por comas, lo que permite clasificar la película en diferentes categorías.

- **overview (*Resumen*):** Sinopsis o descripción de la película.\
Proporciona una breve sinopsis o descripción de la trama de la película.\
Esta variable es útil para obtener una comprensión rápida de la trama de la película.

- **crew (*Equipo*):** Nombres del equipo de producción.\
Enumera los nombres del equipo de producción involucrados en la creación de la película.\
Esto puede incluir actores, directores, productores y otros miembros clave del equipo.

- **orig_title (*Título Original*):** Título original de la película.\
Nombre original de la película en su idioma de origen.\
Es útil para identificar la película en su versión original.

- **status (*Estado*):** Estado de la película.\
Indica el estado actual de la película.\
Por ejemplo, "Released" significa que la película ya ha sido lanzada, mientras que otros valores podrían indicar que está en preproducción, postproducción, etc.

- **orig_lang (*Idioma Original*):** Idioma original de la película.\
Es el idioma en el que se produjo la película.\
Esto es importante para comprender el idioma en el que se filmó la película antes de cualquier doblaje o subtitulado.

- **budget_x (*Presupuesto*):** Presupuesto de la película.\
Costo estimado de producción de la película.\
Esta variable es esencial para evaluar el aspecto financiero de la película y su relación con los ingresos posteriores.

- **revenue (*Ingresos*):** Ingresos de la película.\
Representan la cantidad total de dinero generada por la película, ya sea en taquilla o a través de otras fuentes de ingresos.\
Esto es crucial para evaluar el éxito financiero de la película.

- **country (*País*):** País de origen de la película.\
Indica el país en el que se produjo la película y puede tener implicaciones culturales y de mercado.

```python
# Detectar 'Valores Atípicos' (calcular 'rango intercuartil' (IQR))
dataMovies.describe()
```

```python
# Calcular el rango intercuatílico (IQR) para la columna "Score"
Q1=dataMovies["score"].quantile(0.25)
Q3=dataMovies["score"].quantile(0.75)
IQR= Q3-Q1

lower_bound= Q1 - 1.5 * IQR
upper_bound= Q3 + 1.5 * IQR

outliers= dataMovies[(dataMovies["score"]<lower_bound) | (dataMovies["score"]> upper_bound)]

print("Valores Atipicos:")
print(outliers)
```

```python
# Se crea un gráfico de caja (boxplot) con los datos anteriores que muestra los 'valores atípicos'
plt.figure(figsize=(18, 8))
sns.boxplot(x=dataMovies["score"], color="skyblue", flierprops=dict(markerfacecolor='r', marker='D'), linewidth=2)
plt.title('Distribución de Puntuación (Score) con Valores Atípicos')
plt.xlabel('Puntuación (Score)')
plt.show()
```

```python
# Configuración del 'estilo' de los gráficos
sns.set(style="whitegrid")

# 1 - names (Nombres) - Gráfico de 'Barras Horizontales'
plt.figure(figsize=(12, 6))
sns.countplot(data=dataMovies, y='names', order=dataMovies['names'].value_counts().index[:10])
plt.title('Distribución de Nombres de Películas (names)')
plt.xlabel('Número de Películas')
plt.ylabel('Nombre de la Película')
plt.show()

# Se selecciona un 'subconjunto' aleatorio de filas (por ejemplo de 10 filas) para agilizar la muestra de datos
n_muestras = 10
subconjunto = dataMovies.sample(n=n_muestras, random_state=42)
# 2 - date_x (Fecha) para el 'subconjunto' seleccionado - 'Histograma'
plt.figure(figsize=(10, 5))
sns.histplot(data=subconjunto, x='date_x', bins=30, kde=True)
plt.title('Distribución de Fechas de Lanzamiento (date_x) - Muestra Aleatoria')
plt.xlabel('Fecha de Lanzamiento')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()

# 3 - score (Puntuación) - 'Histograma'
plt.figure(figsize=(10, 5))
sns.histplot(data=dataMovies, x='score', bins=30, kde=True, color='orange')
plt.title('Distribución de Puntuación (score)')
plt.xlabel('Puntuación')
plt.ylabel('Frecuencia')
plt.axvline(dataMovies['score'].mean(), color='red', linestyle='dashed', linewidth=2, label='Media')
plt.show()
# Se puede observar en el histograma que las votaciones más frecuentes se encuentran en los intervalos de 60 y 70 aproximadamente
# Siendo el primero de los dos enumerados el de mayor cantidad de casos.
# La distribución asociada al histograma tiene una asimetría en torno a valores más grandes, además de que se puede visualizar
# que la media esta localizada en 63,5.
# Otra característica a destacar de la gráfica es el primer intervalo entre 0 y 3 aproximadamente, que tiene mucha mayor frecuencia
# respecto de otros intervalos circundantes no tan pronunciados como los de 40 en adelante. Entre 86 en adelante, las frecuencias son muy bajas.

# 4 - genre (Género) - Gráfico de 'Barras Vertical'
plt.figure(figsize=(10, 6))
sns.countplot(data=dataMovies, x='genre', order=dataMovies['genre'].value_counts().index[:10])
plt.title('Distribución de Géneros Cinematográficos (genre) - Top 10')
plt.xlabel('Género')
plt.ylabel('Número de Películas')
plt.xticks(rotation=45)  # Se rotan las etiquetas del eje 'x' para una mayor claridad
plt.show()
# Se visualiza un top 10 de los géneros de películas, siendo el más destacado el drama y el menos destacado del top, el género romance.

# 5 - overview (Resumen)
# No se puede realizar un análisis univariado para datos de texto sin procesar.

# 6 - crew (Equipo)
# No se puede realizar un análisis univariado para datos de texto sin procesar.

# 7 - orig_title (Título Original)
# No se puede realizar un análisis univariado para datos de texto sin procesar.

# 8 - status (Estado) - Gráfico de 'Barras Horizontales'
plt.figure(figsize=(10, 5))
sns.countplot(data=dataMovies, y='status', order=dataMovies['status'].value_counts().index[:10])
plt.title('Distribución de Estados de las Películas (status)')
plt.xlabel('Número de Películas')
plt.ylabel('Estado')
plt.show()
# Este dataset cuenta principalmente con películas lanzadas y que no se encuentran produccion / post produccion

# 9 - orig_lang (Idioma Original) - Gráfico de 'Torta' o 'Circular' ('Pie Chart')
plt.figure(figsize=(8, 8))
orig_lang_counts = dataMovies['orig_lang'].value_counts()[:5]  # Se toman sólo las 5 principales categorías
plt.pie(orig_lang_counts, labels=orig_lang_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Idiomas Originales (orig_lang)')
plt.show()
# En este gráfico visualizamos en un gráfico de torta el porcentaje de los 5 lenguajes originales con los cuales las películas son filmadas

# 10 - budget_x (Presupuesto) - Gráfico de 'Línea'
plt.figure(figsize=(20, 5))
# Se utiliza la función 'plot' para trazar una línea
plt.plot(dataMovies['budget_x'], marker='o', linestyle='-', markersize=2, color='blue')
plt.title('Distribución de Presupuesto de Producción (budget_x)')
plt.xlabel('Índice de Película')
plt.ylabel('Presupuesto')
plt.grid(True)  # Se agrega una 'cuadrícula'
plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(data=dataMovies, x='budget_x', bins=10, kde=True, color='blue')
plt.title('Distribución de Puntuación (presupuesto)')
plt.xlabel('Presupuesto')
plt.ylabel('Frecuencia')
plt.axvline(dataMovies['budget_x'].mean(), color='red', linestyle='dashed', linewidth=2, label='Media')
plt.show()
# Primeramente observamos que ambos gráficos se encuentran en amplitudes (presupuestos) de cientos de millones de dolares.
# Podemos observar el presupuesto de dos maneras, tanto como un gráfico de línea en el cual vemos que las mismas se encuentran
# Muy similarmente distribuidas en torno a valores iguales o inferiores a cien millones de dolares o en un histograma que resulta asimétrico
# hacia los intervalos que adoptan valores iguales o inferiores a cien millones de dolares con un decremento en la frecuencia hacia los mayores presupuestos.

# 11 - revenue (Ingresos) - Gráfico de 'Radar' ('Radar Chart' o 'Spider Plot')
# Se calculan las 'estadísticas' de ingresos (por ejemplo: mínimo, máximo y promedio)
min_income = dataMovies['revenue'].min()
max_income = dataMovies['revenue'].max()
avg_income = dataMovies['revenue'].mean()
# Para crear un 'Gráfico de Radar'
categories = ['Mínimo', 'Máximo', 'Promedio']
values = [min_income, max_income, avg_income]
# Se agrega la primera categoría al final para cerrar el gráfico
categories += [categories[0]]
values += [values[0]]
# Se calcula el ángulo para cada categoría
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
# Se cierra el gráfico
values += values[:1]
angles += angles[:1]
# Se crea el 'Gráfico de Radar'
plt.figure(figsize=(8, 8))
plt.polar(angles, values, marker='o', linestyle='-')
plt.fill(angles, values, alpha=0.25)
plt.title('Distribución de Ingresos Generados (revenue)')
plt.xticks(angles[:-1], categories)
plt.show()

# 12 - country (País) - Gráfico de 'Burbujas' ('Bubble Chart')
# Se seleccionan los 10 países principales
top_countries = dataMovies['country'].value_counts().index[:10]
# Se crea un 'DataFrame' con el número de películas de estos países
country_data = dataMovies[dataMovies['country'].isin(top_countries)]
country_counts = country_data['country'].value_counts()
# Se calcula el tamaño de las burbujas proporcional al número de películas
bubble_size = country_counts.values
# Para crear el 'Gráfico de Burbujas'
plt.figure(figsize=(12, 8))
plt.scatter(
    x=top_countries,  # Países
    y=[1] * len(top_countries),  # Altura constante para posicionar las burbujas en la misma línea
    s=bubble_size,  # Tamaño de las burbujas (proporcional al número de películas)
    alpha=0.5,
    c='pink',  # Color de las burbujas
    edgecolors='gray',  # Color del borde de las burbujas
    linewidth=1  # Grosor del borde de las burbujas
)
plt.title('Gráfico de Burbujas: Número de Películas por País')
plt.xlabel('País')
plt.ylabel('Altura')
plt.xticks(rotation=45)  # Se rotan las etiquetas del eje 'x' para mayor claridad
# Etiquetas para las burbujas
for i, txt in enumerate(bubble_size):
    plt.annotate(txt, (top_countries[i], 1), fontsize=10, ha='center')
plt.show()
```

Realizar análisis de matriz correlación y explicar que variable están correlacionadas.

```python
# Desactivar las 'advertencias'
import warnings
warnings.filterwarnings("ignore")
# Volver a activar las 'advertencias'
#warnings.filterwarnings("default")

#correlacion lineal
correlacion= dataMovies.corr()
print(correlacion)
```

```python
#grafico de correlacion con Seaborn
sns.heatmap(correlacion, annot=True, cmap="YlGnBu" , vmax=1, vmin=-1)
plt.title("Grafico de Calor")
plt.show()

# vemos la correlación entre las variables numericas.
# Notamos primeramente una correlación negativa débil entre el presupuesto y el porcentaje. Siendo que si el puntaje disminuye,
# el presupuesto aumenta o al revés. No hay correlación significativa entre las ganancias (revenue) y el puntaje.
# Por último, existe una correlación positiva entre el presupuesto y las ganancias que la película genera. Indicando que a mayor
# presupuesto, mayores ganancias o a menor presupuesto, las películas reportan menor ganancias.

## Segunda versión utilizando el año de 'date_x'
# Se extrae sólo el año de la columna 'date_x'
dataMovies['year'] = pd.to_datetime(dataMovies['date_x']).dt.year
# Se calcula la 'matriz de correlación' usando los años en lugar de fechas completas
correlation = dataMovies.corr()
# Se grafica la correlación con Seaborn
sns.heatmap(correlation, annot=True, cmap="rainbow", vmax=1, vmin=-1) # En esta versión se utiliza la paleta 'rainbow'
plt.title("Gráfico de Calor (2)")
plt.show()
```

**Explicación del Gráfico o Matriz de 'Dispersión':**
